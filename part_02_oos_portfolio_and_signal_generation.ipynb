## Johansen Test Setup

This is similar to that of part 1, but I will modify this slightly to better fit our needs.

```python
def get_johansen(X, det_order=0, k_ar_diff=1):
    """
    Get the cointegration vectors at 95% level of significance
    given by the trace statistic test.
    
    Parameters
    ----------
    endog : array-like (nobs_tot x neqs)
        The data with presample.
    det_order : int
        * -1 - no deterministic terms
        * 0 - constant term
        * 1 - linear trend
    k_ar_diff : int, nonnegative
        Number of lagged differences in the model.
    """

    N, l = X.shape
    jres = coint_johansen(X, def_order, k_ar_diff)
    trstat = jres.lr1                       # trace statistic
    tsignf = jres.cvt                       # critical values

    for i in range(l):
        if trstat[i] > tsignf[i, 1]:     # 0: 90%  1:95% 2: 99%
            r = i + 1
    jres.r = r
    jres.evecr = jres.evec[:, :r]

    return jres
```

### Kalman Filter

We will use a kalman filter on the mr portfolio price series before we calculate the z-score for the distance from the rolling mean

```python
# TODO: adjust this function to take in hedge ratios and price df
# TODO: will need to look ahead by max(std_window, mav_window) in order to have a value
# for every period
# TODO: desired output is either just a z-score factor or an additional
# kalman filtered price series (used to produce the z-score)
def get_kal_filtered_spreads_and_z_scores(k_beta_spreads, std_window=30, mav_window=30):
    kalman_spreads = k_beta_spreads.copy()
    kal_spread_z_scores = k_beta_spreads.copy()
    
    for pair in k_beta_spreads.columns:
        spread = k_beta_spreads[pair]
        
        # Construct a Kalman Filter to filter spread
        kf = KalmanFilter(transition_matrices = [1],
                          observation_matrices = [1],
                          initial_state_mean = 0,
                          initial_state_covariance = 1,
                          observation_covariance = 1,
                          transition_covariance = 0.1)

        # Use the observed values of the price to get a rolling mean
        state_means, _ = kf.filter(spread.values)

        kalman_spread = pd.Series(state_means.flatten(), index=spread.index)
        # Get the moving average
        # used as mean to calculate rolling z-scores
        spread_mavg = spread.rolling(window=mav_window).mean()
        
        std_rolling = spread.rolling(window=std_window).std()
        
        zscore_kalman_spread = (kalman_spread - spread_mavg)/std_rolling
        
        kalman_spreads[pair] = kalman_spread
        kal_spread_z_scores[pair] = zscore_kalman_spread
        
    return kalman_spreads, kal_spread_z_scores
```
